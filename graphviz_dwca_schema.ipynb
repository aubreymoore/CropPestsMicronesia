# Generate a DwCA Schema Using GraphViz

### Useful Reference for GraphViz
http://www.graphviz.org/doc/info/shapes.html

```python
%%writefile dwca_schema.dot

digraph g { 
    
rankdir = "RL"

node [shape=plaintext] 
    
    taxon [label=<<table BORDER="0" CELLBORDER="1" CELLSPACING="0">
        <tr><td bgcolor="#D8D8D8">taxon<br/>http://rs.tdwg.org/dwc/terms/Taxon</td></tr>
        <tr><td port="ID">ID</td></tr>
        <tr><td>scientificName<br/>http://rs.tdwg.org/dwc/terms/scientificName</td></tr>
        <tr><td>taxonRank<br/>http://rs.tdwg.org/dwc/terms/taxonRank</td></tr>
        <tr><td>kingdom<br/>http://rs.tdwg.org/dwc/terms/kingdom</td></tr>
        <tr><td>bibliographicCitation<br/>http://purl.org/dc/terms/bibliographicCitation</td></tr>
        </table>>]; 
    
    distribution [label=<<table BORDER="0" CELLBORDER="1" CELLSPACING="0">
        <tr><td  bgcolor="#D8D8D8">distribution<br/>http://rs.gbif.org/terms/1.0/Distribution</td></tr>
        <tr><td port="ID">ID</td></tr>
        <tr><td>locality<br/>http://rs.tdwg.org/dwc/terms/locality</td></tr>
        <tr><td>occurrenceRemarks<br/>http://rs.tdwg.org/dwc/terms/occurrenceRemarks</td></tr>
        </table>>]; 
    distribution:ID -> taxon:ID
    
    vernacular [label=<<table BORDER="0" CELLBORDER="1" CELLSPACING="0">
        <tr><td  bgcolor="#D8D8D8">vernacular<br/>http://rs.gbif.org/terms/1.0/VernacularName</td></tr>
        <tr><td port="ID">ID</td></tr>
        <tr><td>vernacularName<br/>http://rs.tdwg.org/dwc/terms/vernacularName</td></tr>
        </table>>]; 
    vernacular:ID -> taxon:ID

    associates [label=<<table BORDER="0" CELLBORDER="1" CELLSPACING="0">
        <tr><td  bgcolor="#D8D8D8">associates<br/>http://rs.tdwg.org/dwc/terms/ResourceRelationship</td></tr>
        <tr><td port="ID">ID</td></tr>
        <tr><td>relationshipOfResource<br/>http://rs.tdwg.org/dwc/terms/relationshipOfResource</td></tr>
        <tr><td port="relatedResourceID">relatedResourceID<br/>http://rs.tdwg.org/dwc/terms/relatedResourceID</td></tr>                           
        </table>>]; 
    associates:ID -> taxon:ID
    associates:relatedResourceID -> taxon:ID

} 
```

```python
# Generate dwca_schema.dot.png and dwca_schema.dot.pdf.
# graphviz must be installed on your system for this to work
!dot -Tpng -O dwca_schema.dot
!dot -Tpdf -O dwca_schema.dot
```
