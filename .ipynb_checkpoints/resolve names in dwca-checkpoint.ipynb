# Resolving names in a Darwin Core Archive

https://github.com/BelgianBiodiversityPlatform/python-dwca-reader

http://links.gbif.org/checklist_best_practices

From text to structured data: Converting a word-processed floristic checklist into Darwin Core Archive format
https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3281575/

* The objective is to verify that each scientificName in a DwC-A is correctly spelled and is the current accepted name for a taxon. 
* Input is a DwC-A which passes the GBIF [DwC-A validator](http://tools.gbif.org/dwca-validator/).
* Only the core table is modified.

### Processing Steps
1. Read core table.
2. Use Global Names Architecture (GNA) to resolve names.
3. Update core table using JSON data returned by the GNA

### Original Core Table

field name|description
---|---
uuid|Universal unique identifier
scientificName|Scientific name exactly how it appears in the document
taxonRank|Taxonomic rank of scientificName; usually **genus** or **species**
kingdom|Usually **Animalia** or **Plantae**

### Core Table with Resolved Names
field name|description
---|---
uuid|Universal unique identifier
scientificName|Scientific name exactly how it appears in the document
taxonRank|Taxonomic rank of scientificName; usually **genus** or **species**
kingdom|Usually **Animalia** or **Plantae**
scientificNameID|Link to GBIF data for this name, equal to acceptedNameUsageID if this is the accepted name for the taxon
acceptedNameUsageID|Link to GBIF data for the accepted name
acceptedNameUsage|Accepted scientificName

```python
from dwca.read import DwCAReader
import pandas as pd
import requests
from ipywidgets import FloatProgress
from IPython.display import display
from os.path import isfile
import json
from datetime import datetime
```

```python
# Step 1. Read core table.

dwca = DwCAReader('testdwca.zip')
core_file_path = dwca.absolute_temporary_path(dwca.descriptor.core.file_location)
df_core = pd.read_csv(core_file_path, index_col=0)
pd.options.display.max_rows = 4
df_core
```

```python
def timestamp():
    return datetime.now().strftime("%Y%m%d-%H%M%S")
    
#timestamp()
```

```python
%%time

# Step 2. Use Global Names Architecture (GNA) to resolve names.

# If there are more than 1000 names, there will be a problem.

# Build data string
data_str = ''
for index, row in df_core.iterrows():
    data_str += '{}|{}\n'.format(index, row.scientificName)
data_str = data_str[:-1] # Remove final newline

# Resolve names

print('Accessing "reslover.globalnames.org" to resolving {} taxonomic names...'.format(df_core.shape[0]))
params = {'data': data_str, 
          'preferred_data_sources': '11', 
          'best_match_only': 'true',
          'with_context': 'true',
          'with_canonical_ranks': 'false'
         }
url = 'http://resolver.globalnames.org/name_resolvers.json'
r = requests.post(url, params)
file_name = 'resolved_names_{}.json'.format(timestamp())
resolved_names_json = r.json()
json.dump(resolved_names_json, open(file_name,'w'), indent=2)
print('Results written to "{}".'.format(file_name))
```

```python
mylist = []
for data_dict in resolved_names_json['data']:
    
    # Get top level items that are not lists.
    
    toplevel_dict = {}
    for key, value in data_dict.items():
        if not type(value) is list:
            toplevel_dict.update({key: value})
            
    # Get items in the "preferred_results" list.
    # If this list does not exist, get items from the "results" list instead.
    
    if 'results' in data_dict:
        if data_dict['preferred_results'] == []:
            results_list = data_dict['results']
        else:
            results_list = data_dict['preferred_results']
        if len(results_list) == 1 and results_list[0]:
            results_dict = results_list[0]
    merged = {**toplevel_dict, **results_dict}                       
    mylist.append(merged)
    
# Import into a pandas data frame and write to a CSV file

df = pd.DataFrame(mylist)
df.set_index('supplied_id', inplace=True)
pd.options.display.max_columns = 1000
pd.options.display.max_colwidth = 1000
filename = 'species_info_{}.csv'.format(timestamp())
df.to_csv(filename, index=False)
print('Data written to ', filename)
df.transpose() 
```

```python
def lineage(classification_path_ranks, classification_path):
    """
    Given:
        classification_path_ranks = "kingdom|phylum|class|order|family|genus|species"
        classification_path = "Plantae|Tracheophyta|Liliopsida|Asparagales|Asparagaceae|Agave|Agave sisalana"        
    Returns:
        {'class': 'Liliopsida',
        'family': 'Asparagaceae',
        'genus': 'Agave',
        'kingdom': 'Plantae',
        'order': 'Asparagales',
        'phylum': 'Tracheophyta',
        'species': 'Agave sisalana'}
    """
    return dict(zip(classification_path_ranks.split('|'), classification_path.split('|')))
```

```python
mylist = []
for data_dict in resolved_names_json['data']:
    mydict = {}
    mydict['uuid'] = data_dict['supplied_id']
    mydict['supplied_name_string'] = data_dict['supplied_name_string']
    mydict['is_known_name'] = data_dict['is_known_name']
    if 'results' in data_dict:
        if data_dict['preferred_results'] == []:
            results_list = data_dict['results']
        else:
            results_list = data_dict['preferred_results']
        if len(results_list) == 1 and results_list[0]:
            results_dict = results_list[0]
            mydict['scientificNameID'] = results_dict.get('taxon_id','')
            mydict['scientificName'] = results_dict.get('name_string','')
            mydict['data_source_title'] = results_dict.get('data_source_title','')
            
            # Handle synonomy
            if 'current_taxon_id' in results_dict:
                mydict['acceptedScientificNameID'] = results_dict.get('current_taxon_id','')
            else:
                mydict['acceptedScientificNameID'] = mydict.get('scientificNameID','')
                
            if 'current_name_string' in results_dict:
                mydict['acceptedScientificName'] = results_dict.get('current_name_string','')
            else:
                mydict['acceptedScientificName'] = mydict.get('scientificName','')
                
            # Handle classification
            if 'classification_path' in results_dict and 'classification_path_ranks' in results_dict:
                classification_dict = lineage(results_dict['classification_path_ranks'], 
                                              results_dict['classification_path'])
                mydict['kingdom'] = classification_dict.get('kingdom','')
                mydict['phylum'] = classification_dict.get('phylum','')
                mydict['class'] = classification_dict.get('class','')
                mydict['order'] = classification_dict.get('order','')
                mydict['family'] = classification_dict.get('family','')
                mydict['genus'] = classification_dict.get('genus','')
                mydict['species'] = classification_dict.get('species','')
            
    mylist.append(mydict)
df = pd.DataFrame(mylist)
df.set_index('uuid', inplace=True)
pd.options.display.max_rows = 1000
pd.options.display.max_colwidth = 1000
df.to_csv('species_info.csv', index=False)
df.transpose() 
```
